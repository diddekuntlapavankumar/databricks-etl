```python
import csv

def read_csv(file_name):
    with open(file_name, 'r') as file:
        reader = csv.reader(file)
        next(reader)  # skip header
        return list(reader)

def filter_age(data):
    result = []
    for row in data:
        if int(row[2]) > 25:
            result.append(row[1])  # name
    return result

def main():
    data = read_csv('/Volumes/workspace/default/source/customers1.csv')
    names = filter_age(data)
    
    for name in names:
        print(name)

main()
```

```python
def read_customers_csv(path):
    return spark.read.csv(path, header=True, inferSchema=True)

def filter_customers_by_age(df, min_age):
    return df.filter(df['age'] > min_age).select('name')

def main():
    df = read_customers_csv('/Volumes/workspace/default/source/customers1.csv')
    filtered_df = filter_customers_by_age(df, 25)
    display(filtered_df)

main()
```

```python
import unittest

class TestCSVProcessing(unittest.TestCase):
    @classmethod
    def setUpClass(cls):
        cls.sample_data = [
            ('1', 'Pavan', 25, 'Bangalore'),
            ('2', 'Ravi', 30, 'Hyderabad'),
            ('3', 'Amit', 22, 'Delhi')
        ]
        cls.columns = ["id", "name", "age", "city"]
        cls.df = spark.createDataFrame(cls.sample_data, cls.columns)

    def test_read_customers_csv(self):
        df = read_customers_csv('/Volumes/workspace/default/source/customers1.csv')
        self.assertIn('id', df.columns)
        self.assertIn('name', df.columns)
        self.assertIn('age', df.columns)
        self.assertIn('city', df.columns)

    def test_filter_customers_by_age(self):
        filtered = filter_customers_by_age(self.df, 25)
        names = [row['name'] for row in filtered.collect()]
        self.assertIn("Ravi", names)
        self.assertNotIn("Pavan", names)
        self.assertNotIn("Amit", names)

if __name__ == "__main__":
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
