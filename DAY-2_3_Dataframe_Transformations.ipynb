```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, sum, row_number
from pyspark.sql.window import Window

spark = SparkSession.builder.getOrCreate()

data = [
    (1, "Pavan", "IT", 5000),
    (2, "Ravi", "HR", 4000),
    (3, "Amit", "IT", 6000),
    (4, "John", "HR", 3000),
    (5, "Sara", "IT", 7000)
]

df = spark.createDataFrame(data, ["id", "name", "department", "salary"])
display(df)
```

```python
df1 = df.withColumnRenamed("salary", "emp_salary")
```

```python
df2 = df1.withColumn("salary_bonus", col("emp_salary") + 1000)
display(df2)
```

```python
df3 = df2.agg(sum("emp_salary").alias("total_salary"))
display(df3)
```

```python
df4 = df2.groupBy("department").agg(sum("emp_salary").alias("dept_total_salary"))
display(df4)
```

```python
df5 = df4.orderBy("dept_total_salary", ascending=False)
display(df5)
```

```python
window_spec = Window.partitionBy("department").orderBy(col("emp_salary").desc())
df6 = df2.withColumn("rank", row_number().over(window_spec))
display(df6)
```

```python
df2.write.format("delta") \
  .partitionBy("department") \
  .mode("overwrite") \
  .save("/Volumes/workspace/default/source/employee")
```
