```python
class MultiLineJsonReader:
    def __init__(self, file_path):
        self.file_path = file_path

    def read(self):
        return spark.read.option("multiLine", True).json(self.file_path)

# Usage
reader = MultiLineJsonReader('/Volumes/workspace/default/source/databricks_json_Samples/employees_multiline.json')
df = reader.read()
display(df)
```

```python
class ComplexJsonReader:
    def __init__(self, file_path):
        self.file_path = file_path

    def read(self):
        return spark.read.option("multiLine", True).json(self.file_path)

# Usage
reader = ComplexJsonReader('/Volumes/workspace/default/source/databricks_json_Samples/orders_complex.json')
df = reader.read()
display(df)
```
