```python
# Create a sample DataFrame
df = spark.createDataFrame([(1, "example"), (2, "data")], ["id", "value"])

# Write a Spark DataFrame to a Unity Catalog table
df.write.format("delta").mode("overwrite").saveAsTable("workspace.default.table_name")
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("value", StringType(), True)
])

df = spark.createDataFrame([(1, "example"), (2, "data")], schema)
```
